## For Valid

```python
import os
import pandas as pd

# 1. SETUP - Use your folder name
BASE_DATA_DIR = "Major Project Dataset" 
OUTPUT_CSV = "master_valid_dataset.csv"

languages = ['bengali', 'gujarati', 'hindi', 'kannada', 'malayalam', 
             'marathi', 'odia', 'punjabi', 'sanskrit', 'tamil', 'telugu', 'urdu']

master_data = []

print("🚀 Starting Extension-Aware CSV Creation...")

for lang in languages:
    transcript_path = os.path.join(BASE_DATA_DIR, lang, "valid", "transcription_n2w.txt")
    audio_folder = os.path.join(BASE_DATA_DIR, lang, "valid", "audio")

    if not os.path.exists(transcript_path) or not os.path.exists(audio_folder):
        continue

    print(f"📂 Processing: {lang}...")

    # A. Load Transcripts
    lang_transcripts = {}
    with open(transcript_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            parts = line.split() 
            if len(parts) >= 2:
                # Corrected the typo here
                file_id_in_text = parts[0].strip()
                text = " ".join(parts[1:]).strip()
                lang_transcripts[file_id_in_text] = text

    # B. Match Audio Files
    audio_files = [f for f in os.listdir(audio_folder) if f.lower().endswith((".wav", ".m4a"))]
    match_count = 0
    
    for file_name in audio_files:
        # Match against the exact filename (e.g., '844424930403298-711-f.m4a')
        if file_name in lang_transcripts:
            master_data.append({
                "file_path": os.path.abspath(os.path.join(audio_folder, file_name)),
                "language": lang,
                "transcription": lang_transcripts[file_name],
                "file_id": file_name
            })
            match_count += 1
    
    if match_count > 0:
        print(f"   ✅ Matched {match_count} files.")

# 2. SAVE & SUMMARY
if master_data:
    df = pd.DataFrame(master_data)
    df.to_csv(OUTPUT_CSV, index=False, encoding="utf-8-sig")
    print(f"\n✨ SUCCESS! Saved {len(df)} samples to {OUTPUT_CSV}")
    print("-" * 30)
    print(df['language'].value_counts())
    
    print("\n📝 Preview (First 5 rows):")
    print(df[['language', 'file_id', 'transcription']].head())
else:
    print("\n❌ Still 0 samples. Double-check if the file names in the audio folder EXACTLY match the first column in your .txt file.")
```
