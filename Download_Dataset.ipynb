```python
"""
Optimized Google Drive File Downloader

Downloads a specific file from Google Drive with speed optimizations.
"""


import os
import sys
import argparse

try:
    import gdown
except ImportError:
    print("Error: gdown library not found.")
    print("Install or upgrade it using: pip install --upgrade gdown")
    sys.exit(1)

# File to download
DRIVE_FILE_ID = "17-Gs4NQvG405k-q_ewKc3Q-lBhnf6Dsa"

def download_file(output_path=None):
    """
    Download the file from Google Drive using high-speed mode.
    """
    try:
        # If no output path is given, gdown uses the original filename from Drive
        if output_path:
            print(f"Target destination: {output_path}")
            output_dir = os.path.dirname(output_path)
            if output_dir and not os.path.exists(output_dir):
                os.makedirs(output_dir, exist_ok=True)
        else:
            print("Downloading to current directory...")

        # OPTIMIZATIONS:
        # 1. Use the ID directly rather than the UC URL string for better resolution.
        # 2. 'fuzzy=True' helps gdown bypass certain Google Drive permission/proxy redirects.
        # 3. 'speed' is generally limited by Google's server-side throttling, 
        #    but ensuring we aren't using a proxy helps.
        
        output = gdown.download(
            id=DRIVE_FILE_ID, 
            output=output_path, 
            quiet=False, 
            fuzzy=True
        )
        
        if output:
            print(f"\n✓ Download complete: {output}")
            return True
        else:
            print("\n✗ Download failed.")
            return False
            
    except Exception as e:
        print(f"\n✗ Error: {str(e)}")
        # If it fails, suggest a cache clear which often fixes 'slow' or 'stuck' gdown sessions
        print("Tip: If download is stuck, try running 'gdown --clear-cache' in your terminal.")
        return False

def main():
    """Main function with Jupyter compatibility"""
    parser = argparse.ArgumentParser(description='Fast Dataset Downloader')
    parser.add_argument('-o', '--output', help='Output file path', default=None)
    
    # parse_known_args prevents crashing in Jupyter Notebooks/VS Code
    args, unknown = parser.parse_known_args()
    
    print("=" * 40)
    print(f"INITIATING DOWNLOAD: {DRIVE_FILE_ID}")
    print("=" * 40)
    
    success = download_file(args.output)
    return 0 if success else 1

if __name__ == '__main__':
    try:
        status = main()
        if status != 0:
            sys.exit(status)
    except SystemExit:
        pass
```
